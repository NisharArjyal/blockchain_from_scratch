### Cryptographic Hash Function(chf)

chf = math fn. with properties: 

1. input can be string of any size
2. produces fixed sized output, assume 256-bit output
3. efficiently computable, this means for given string, output of hash fn is in reasonable amount of time. Computing hash of n-bit string should be $\mathcal{O}(n)$

such hash fn can be used to build a data str such as a hash table. 

only focus on chf for now: 

for hash fn to be cryptographically secure : 
must have folowing prprties

1. collision resistance 
2. hiding
3. puzzle friendliness 

**collision resistance** : 

collision occurs when two distinct inputs produce same output

hash fn s collision resistant if noone can find a collision (this is diffn from saying no collision exists) 

A hash function H is said to be collision
resistant if it is infeasible to find two values, x and y, such that
x $\neq$ y, yet H(x) = H(y).

collisions exist for any hash fn simply from a counting arguement

this is because the input space is larger than the output space. 

**hiding**:

asserts that if we're given the output of hash fn, there's no feasible way to figure out what input $x$ was. 

> A hash fn H is said to be hiding if when a secret value r is chosen from a probability distribution that has high min-entropy, then, given $H(r\parallel x)$, it is infeasible to find x.  

min entropy = measure of how predictable outcome is and high min-entropy says distribn is spread out. 

application of hiding: 

_commitment_: digital analog of taking a value, sealing in envelope, and putting envelope on table where everyone can see

doing this, you've committed to what's inside envelope
However, you've not opened it, so even though you've committed to value, value remains secret from everyone else

_Commitment Scheme_: consists of two algos: 

1. _com_:= commit(msg, nonce) commit fn takes msg and secret random value, called nonce as input and returns commitment
2. verify:= (com, msg, nonce) verify fn takes commitment, secret random value (nonce) and msg as input. It returns true if com == commit(msg, nonce) and false otherwise 

The following two security properties have to hold however : 

**Hiding**: given _com_, it is infeasible to find msg
**Binding**: infeasible to find 2 pairs (msg, nonce) and (msg', nonce') such that $msg\neq msg\prime$ and commit(msg, nonce) $==$ commit(msg$\prime$, nonce$\prime$)

in crypto _nonce_ = value that can only be used once 

these two properties behave like someone sealing and opening at envelope

First given _com_, the commitment, some1 looking at envelope can't figure what the msg is, 2nd prprty is that it's binding. ensures that when u commit to what's in envelope, u can't change mind later

> How to know that these properties hold? 

by implementing a commitment scheme

do so using crypto hash fn 

commit(msg, nonce) := H($nonce \parallel msg$), wherer nonce = random value 256 bit 

to commit to a msg, generate random 256 bit nonce, then concatenate nonce and msg and return the hash of concatenated value as commitment


NOTE: if H is a hash fn that is both collision resistant and hiding, this commitment scheme will work because collision resistance is an equivalent way of looking at binding

**Puzzle Friendliness**: 

> A hash fn $H$ is said to be puzzle friendly if for every possible n-bit output value y, if k is chosen from a distrbn with high min-entropy, then it is infeasible to find x such that $H(k \parallel y)=y$ in time significantly less than $2^{n}$


meaning: 
if someone wants to target hash fn to have some particular output vlaue y, and if part of input has been chosen in suitably randomized way, then very difficult to find another value that hits target 

#### SHA-256 : 

bitcoin uses this hash fn, pretty good to use 

hash fns work on inputs of arbitrary lengths and produce outputs of fixed length.

There's a generic method so that as long as hash fn works on fixed-length inputs can be converted to hash fn that works on arbitrary length inputs. called the **Merkle-Damgard** transform. SHA-256 is one of no. of commonly used hash fn tha make use of this method. 

fixed-length collision resistant hash fn is called _compression_ fn. if compression fn is collision resistant, then the created hash fn will also be collision resistant. 

Consider hash fn input of length m and output n < m

inputs (can be of any size) divided into blocks of length (m-n) 
pass each block together with output of previous block into compression function 
so input length = m-n + n = m, which used to be input of compression fn 
for the case of no previous block output, use preset _initialization vector_ (IV). 

this IV can be looked up in a standard's document

last block's output = result that we last return 

this puzzle friendliness seems to be of importance only in bitcoin mining and not generally used in hashing in other contexts

### Hash Pointer: 

1. Data str

1. regular ptr gives  a way to retrieve infmtn, hash ptr also allows to verify that the infmtn hasn't been changed 

can use hash ptrs to build many types of D.Strs. like linked list, binary search tree, and implement with hash pointers instead of ordinary pointers as we normally would. 

A linked list using hash pointers = block chain 
in a block chain, each block not only tells where value of previous block was but also contains a digest of that value, which allows to verify that value hasn't changed.
